```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#O que já sabemos?
empresas =['Apple', 'Google','Amazon','Microsoft','Tencent',"McDonald's",'Visa','Facebook','Alibaba', 'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646, 611.460, 214.023, 196.526, 191.032, 186.421,169.966, 124.273]
```

```python
#Já é possível plotar, só com as variáveis 
plt.plot(empresas,bilhoes)
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#Redimensionando 
plt.rcParams["figure.figsize"] = (15,8)

plt.plot(empresas,bilhoes)
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.title('Empresas mais ricas do mundo')
```

```python
#Conhecendo o fontsize 
plt.rcParams["figure.figsize"] = (15,8)

plt.plot(empresas,bilhoes)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo')
```

```python
#Mudando a cor e legendando 

plt.plot(empresas,bilhoes, color = 'green')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize = 15)
```

```python
#Aumentar a espessura da linha com lw
plt.plot(empresas,bilhoes, color = 'green', lw = 10)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize = 15)
```

```python
#Adcionando marcador 
plt.plot(empresas,bilhoes, color = 'green', marker = 'o')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize = 15)
```

```python
#Adcionando linhas de grade 
plt.plot(empresas,bilhoes, color = 'green', marker = 'o')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize = 15)
plt.grid(True)
```

```python
#Mudando o tipo do gráfico e colocando a legenda acima 

plt.bar(empresas,bilhoes, color = 'green')
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.legend(['Bilhões de Dólares'], loc = 'center right', fontsize = 15)
plt.grid(True)
```

```python
#Gráfico mais simples
plt.bar(empresas,bilhoes)
```

```python
cores = ['black', 'dimgray', 'dimgrey', 'gray', 'grey', 'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey' ]
```

```python
plt.bar(empresas,bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões de dólares', fontsize = 20)
plt.grid(True)
```

```python
plt.style.available
```

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.style.use('seaborn-white')
plt.bar(empresas,bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize = 20)
plt.xlabel('Empresas', fontsize = 20)
```

```python
#barh
plt.style.use('seaborn-dark')
plt.barh(empresas,bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Empresas', fontsize = 20)
plt.xlabel('Bilhões de dólares', fontsize = 20)
plt.show()
```

```python
#Scatter
plt.style.use('seaborn-dark')
plt.scatter(empresas,bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize = 20)
plt.xlabel('Empresas', fontsize = 20)
plt.show()
```
